```python
#Preprocessing an preparation
df_uce.columns
```

```python
df_uce.columns = df_uce.columns.to_series().apply(lambda x: x.strip())
df_uce.columns
```

```python
#Replace white space in column names with underscores
df_uce.columns = df_uce.columns.str.replace(' ', '_')
df_uce.columns
```

```python
#Checking the information on the dataset
df_uce.info()
```

```python
#Check for duplicates in the dataset and drop them
for isduplicate in df_uce.duplicated():
    if isduplicate:
        print("Duplicates identified")
        df_uce.drop_duplicates(inplace=True)
        break
    else:
        pass
print("No duplicates found")
df_uce[df_uce.duplicated()==True]
```

```python
#Check for null and NaN values in the dataset
#df_uce.isna().any()
df_uce.isnull().sum().sort_values(ascending=False)
```

```python
#Check null values in percentage
for i in df_uce.columns:
    if df_uce[i].isnull().any():
        print(i, "-----", df_uce[i].isnull().sum()*100/df_uce.shape[0], "%")
```

```python
#Check missing values in graph
import missingno as msno
msno.bar(df_uce, color = "gray", fontsize =28)
```

```python
#check for the Number of rows that are null
null_rows = df_uce[df_uce.isna().any(axis=1)]
before_computation = len(null_rows.axes[0])

print("Number of rows with null values:" + str(before_computation))
```

```python
#Remove rows without school names
df_uce.dropna(thresh=None, subset=["SCHOOL"], inplace =True)
```
