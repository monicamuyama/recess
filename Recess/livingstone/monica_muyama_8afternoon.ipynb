```python
#Pandas program to compare the elements of two Pandas series. 
#Create the series
series1 = pd.Series([4,65,436,3,9])
series2 = pd.Series([7,0,3,897,9])
compare = series1==series2
comparison = series1.compare(series2)
print(comparison)
print(compare)
```

```python
#Pandas program to add, subtract, multiply and divide two Pandas series.
#Create the series 
series1= pd.Series([2,4,6,8,14])
series2= pd.Series([1,3,5,7,9])

#Addition
addition = series1 + series2
print("Addition:")
print(addition)

# Subtraction
subtraction = series1 - series2
print("\nSubtraction:")
print(subtraction)

# Multiplication
multiplication = series1 * series2
print("\nMultiplication:")
print(multiplication)

# Division
division = series1 / series2
print("\nDivision:")
print(division)
```

```python
#5.	Write a Pandas program to display the most frequent value in the given series and replace everything else as ‘Other’ in the series. 
#(Use the ‘HomeTeamGoals’ column in the dataset provided)
#read the dataset using pandas
df = pd.read_csv('AfricaCupofNationsMatches.csv')

#first view a brief about the dataset and confirm availability of the column
df.head()
```

```python
#Find the most frequent value in the column
most_frequent_value = df['HomeTeamGoals'].mode()[0]

# Replace all other values with 'Other'
df['HomeTeamGoals'] = df['HomeTeamGoals'].apply(lambda x: most_frequent_value if x == most_frequent_value else 'Other')

# Display the modified series
print(df['HomeTeamGoals'])
```
