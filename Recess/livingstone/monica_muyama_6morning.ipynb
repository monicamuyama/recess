```python
#replace using mean, median, mode
#Calculate the mean
m = df4['Calories'].mean()
df4['Calories'].fillna(m, inplace=True)
print(df4.to_string())
```

```python
#Displays all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
df4
```

```python
#Empty cells, wrong format, wrong data, duplicate(core of data prparation and preprocessing)
#calculate MEDIAN
med = df4['Calories'].median()
df4['Calories'].fillna(med, inplace = True)
print(med)
df4
```

```python
#Calculate mode on age
md = df4['Age'].mode()
df4['Age'].fillna(md, inplace = True)
print(md)
df4
```
